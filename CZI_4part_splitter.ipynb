```python
import slideio
import os, glob
from matplotlib import pyplot as plt
import numpy as np
import tifffile as tiff
```

```python
roots = ['/media/sam/PP_20TB_2/Mingxia/2023-07-25/Stitched',
         '/media/sam/PP_20TB_2/Mingxia/2023-07-19',
         ]

# Use the glob function to get the list of file names that match the pattern in the specified folder
pattern = "*.czi" 
file_list = sum([glob.glob(f"{root}/{pattern}") for root in roots],[])
file_list
```

```python
for image_path in file_list:

    print(f"Splitting {image_path}")

    # load the current czi
    slide = slideio.open_slide(image_path, 'CZI')
    scene = slide.get_scene(0)

    # Max project the image
    i=0

    im = np.array([
                np.amax(
                    np.array([scene.read_block(channel_indices=[i], slices = (j,j+1)) 
                    for j in range(4)]), 
                    axis=0) 
                for i in range(4)])
    
    print(f'{image_path} converted to tif')

    c1 = int(im.shape[1]/2)
    c2 = int(im.shape[2]/2)

    im_A = im[:,0:c1,0:c2]
    im_B = im[:,c1+1:,0:c2]
    im_C = im[:,c1+1:,c2+1:]
    im_D = im[:,0:c1,c2+1:]

    tiff.imwrite(image_path[0:-4]+'_A.ome.tif', im_A)
    tiff.imwrite(image_path[0:-4]+'_B.ome.tif', im_B)
    tiff.imwrite(image_path[0:-4]+'_C.ome.tif', im_C)
    tiff.imwrite(image_path[0:-4]+'_D.ome.tif', im_D)
    print(f'{image_path} converted to SEPARATE tif files')
```
